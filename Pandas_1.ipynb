## Pandas Series

Pandas Series is a type that is used to store one column only.
We can think of series as one column of a dataframe extracted.

series is very similar to numpy array, with the major difference being it has an index label for each observation.

```python
df4 = pd.DataFrame(np.random.randint(1,100,(5,4)),columns=list('abcd'))
```

```python
df4
```

```python
df4['a']
```

```python
print(type(df4['a']))
```

```python
# Slicing from pandas series
df4['a'][2]
```

```python
# Creating a list from the slicing elemnets from the series
df4['a'][:4].tolist()
```

## Creating a Standalone Series Object

```python
data = np.arange(10)
index = ['a','b','c','d','e','f','g','h','i','j']

ser = pd.Series(data=data,index=index)
```

```python
ser
```

```python
print(type(ser))
```

```python
# Series are Vectorized Objects by default.
# In order to print squares of every item in series, we need to iterate through loop
# Insted directly perform squaring operation

squares = ser ** 2

print(squares)
```

```python
# Extract Indiviual items from series

ser['g']
```

```python
# To extract more than one item, put all the item labels in a list and then pass it as an arument

ser[['d','i','c']]
```

## Creating Pandas Series from Dictionary

```python
d1 = {'a':1,'b':2,'c':6}
d2 = {'d':8,'f':15,'h':11}

ser1 = pd.Series(d1)
ser2 = pd.Series(d2)
```

```python
print(ser1)
print(ser2)
```

```python
# Adding thesee two series objects

s = ser1.values + ser2.values
```
